```python
for col in df.columns:
  if(df[col].dtype!='object'):
    sns.boxplot(df[col])
    plt.xlabel(col)
    plt.show()
```

```python
#If there is no box in the boxplot, should not consider as outlier. eg- pdays
out_list=['age','campaign','cons.conf.idx']

for col in out_list:
  Q1=df[col].quantile(0.25)
  Q3=df[col].quantile(0.75)
  IQR = Q3-Q1
  lower_bound = Q1 - 1.5*IQR
  upper_bound = Q3 + 1.5*IQR
  df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

```python
# Label Encoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for col in df.columns:
  if(df[col].dtype=='object'):
    df[col]=le.fit_transform(df[col]) #fit-allows to process,transform-to apply changes
```
